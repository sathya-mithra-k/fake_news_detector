```python
import re
def cleaning(text):
    text  = text.lower()
    text = re.sub('[^a-zA-Z]', ' ', text)
    return text
    #start working from this function
    #this should remove unwanted characters, links and other stuffs
    
df['text'] = df['text'].apply(cleaning)
```

```python
#after this i think applying what we have learnt in nlp till now would be helpful to classify even more
#ill be performing tokenization and remove the stop words

import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from string import punctuation

nltk.download('stopwords')
nltk.download('punkt')
def pre_process(text):
    tokens = word_tokenize(text)
    for t in tokens:
        if t not in tokens:
            tokens = t
    return ''.join(text)
```

```python
df['text'] = df['text'].apply(pre_process)
```
